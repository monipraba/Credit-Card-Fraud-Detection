## Naivebayes classification using credit card dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score


dataset=pd.read_csv('creditcard.csv')
```

## Describing Dataset

```python
dataset.describe()
```

```python
dataset.head()
```

## visualizing Dataset

```python
sns.set({'figure.figsize':(5,5)})
sns.countplot(dataset.Class)
print(dataset.Class.value_counts())
```

## Cluster Map

```python
corr=dataset.corr()
sns.clustermap(corr,annot=True,fmt='.2f')
```

## Splitting Training and Test Dataset

```python
target=dataset['Class'].values
train=dataset.drop(['Class'],axis=1).values
x_train,x_test,y_train,y_test=train_test_split(train,target,test_size=0.2)
print("lenth of train :",len(x_train))
print("lenth of test :",len(x_test))
```

## Creating Model

```python
model=GaussianNB()
model.fit(x_train,y_train)
model.score(x_test,y_test)
ypred=model.predict(x_test)
```

## Finding Accuracy of the model

```python
print("accuracy score :",accuracy_score(ypred,y_test))
print("confusion matrix:",confusion_matrix(ypred,y_test))
print("classification report:",classification_report(ypred,y_test))
```
